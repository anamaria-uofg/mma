```python
def plot_peaks_rt(dataframe, title, rt_window, stds = True):
    results_dir = osp+'results/300dpi'
    yaxis = "Total number of aligned peaks"
    p = 0
    if stds:
        p = 1
        yaxis = "Total number of aligned SRM metabolites"
    for n in range(0,6,2):
        s = [eval(i)[p] for i in np.array(dataframe.transpose().iloc[n])]
        print(s)
        s1 = [eval(i)[p] for i in np.array(dataframe.transpose().iloc[n+1])]
        print(s1)
        fig, ax = plt.subplots()
        nice_fonts = {"text.usetex": True,
            "font.family": "serif",
            "font.serif" : "Times New Roman"}
        plt.rcParams.update(nice_fonts)
        
        plt.rc('font', family='serif')
        plt.rc('xtick', labelsize=10)
        plt.rc('ytick', labelsize=10)
        
        ax.plot(rt_window,s, label = 'Unmodified')
        ax.plot(rt_window,s1,label = 'GP modified')
        ax.legend()
        ax.set_ylabel(yaxis, fontsize = 15)
        ax.set_xlabel("JoinAligner RT Window (min)", fontsize = 15)
        
    plt.savefig(os.path.join(results_dir,title+str(n)+'.pdf'),dpi=300)
    plt.show() 
        
    
```

```python
plot_peaks_rt(combinedvl, "vl_standards", rt_window)
```
