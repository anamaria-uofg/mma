```python
def get_ids_with_similar_mz_rt(did, dataset1, dataset2, mz_tol = 0.01,  rt_tol = 5):
    mz = dataset1.loc[did][0]
    rt = dataset1.loc[did][1]
    listid = []
    for i, row in dataset2.iterrows():
        cmz = row['row m/z']
        crt = row['row retention time']
        cid = row['custom_id.1']
        cpv = row['adj-p']

        if cmz > mz-mz_tol and cmz < mz+mz_tol:

            if crt > rt-rt_tol and crt < rt+rt_tol:
                print("{}:Found mz,rt from compound {} within tolerance range:{},{}".format(did,cid, cmz,crt))
                print("mz difference:{}, rt difference:{}".format(cmz-mz, crt-rt))
                print("pval:{}".format(cpv))
                listid.append(cid)
            
            
    return listid
```

```python
def plot_boxplots_individual_dataset(data, metab_id, name_column, ylim = (10,30), y = False ):
    import seaborn as sns
    import matplotlib.pyplot as plt
    plt.figure(figsize=(5,5))

    ax = sns.boxplot(y=data[metab_id], x=name_column, data = data, order = ['controlHATcontrol','infectedHATS1', 'infectedHATS2'])
    ax = sns.swarmplot(y=data[metab_id], x=name_column, data = data, order = ['controlHATcontrol','infectedHATS1', 'infectedHATS2'], color="black")
    if y:
        plt.ylim(ylim)
    plt.show()
```
